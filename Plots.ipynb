```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, num=50)

def f(x, W, B):
    return sigmoid(W*x - B)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, f(x, W=10, B=5))

def update(W = 1, B = 5):
    line.set_ydata(f(x,W,B))
    fig.canvas.draw_idle()
    
interact(update, W = (0,1000,50), B = (0,500,50));
```

```python
%matplotlib notebook

x = np.linspace(0, 1, num=50)

def f(x, a):
    return a*sigmoid(1000*x - 300) - a*sigmoid(1000*x - 400)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, f(x, a=1))

def update(a = 1):
    line.set_ydata(f(x,a))
    fig.canvas.draw_idle()
    
interact(update, a = (0,1,0.1));
```
